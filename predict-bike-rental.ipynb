```python
import pandas as pd

bike = pd.read_csv('bike_rental_hour.csv')
bike
```

```python
bike.describe()
```

```python
bike['cnt'].hist()
```

```python
import seaborn as sns
bike_cols = bike.drop(labels = ['instant','dteday','casual','registered'], axis = 1)
print(bike_cols.corr()['cnt'].sort_values(ascending=False))
sns.heatmap(bike_cols.corr())
```

```python
def assign_label(num):
    if num >= 0 and num < 6:
        return 4
    elif num > 5 and num < 12:
        return 1
    elif num > 11 and num < 18:
        return 2
    elif num > 17 and num <= 24:
        return 3
    
bike['time_label'] = bike['hr'].apply(assign_label)
print(bike[['hr','time_label']])
```

```python
from sklearn.model_selection import train_test_split
bike_model = bike.drop(labels = ['instant','dteday','casual','registered','cnt'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(bike_model, bike['cnt'], test_size=0.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
a0 = lr.intercept_
a1 = lr.coef_
print(a0)
print(a1)

predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,predictions)
mse
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

hyperparameters = {
'min_samples_leaf': range(1,10)
}

tree = DecisionTreeRegressor()
grid = GridSearchCV(tree, param_grid = hyperparameters, cv = 10, scoring = 'neg_mean_squared_error')
grid.fit(X_train, y_train)
best_params = grid.best_params_
best_score = grid.best_score_

print(best_params)
print(best_score)
```

```python
tree = DecisionTreeRegressor(min_samples_leaf = 5)
tree.fit(X_train, y_train)
predictions = tree.predict(X_test)
mse = mean_squared_error(y_test,predictions)
mse
```

```python
from sklearn.ensemble import RandomForestRegressor

tree = RandomForestRegressor(min_samples_leaf = 5)
tree.fit(X_train, y_train)
predictions = tree.predict(X_test)
mse = mean_squared_error(y_test,predictions)
mse
```

```python
tree = RandomForestRegressor()
grid = GridSearchCV(tree, param_grid = hyperparameters, cv = 10, scoring = 'neg_mean_squared_error')
grid.fit(X_train, y_train)
best_params = grid.best_params_
best_score = grid.best_score_

print(best_params)
print(best_score)
```
